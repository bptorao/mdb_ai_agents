## Descripción del agente
El agente recibirá un input del usuario sobre información relacionada a los parkings públicos de Málaga, 
podrá consultar:
clima, ocupación de parkings públicos, horas mas concurridas, hacer predicción sobre ocupación de un parking a cierta hora cierto día

```python
import pandas as pd
import ollama
import requests
from io import StringIO

data_url =  "https://datosabiertos.malaga.eu/api/3/action/datastore_search?resource_id=0dcf7abd-26b4-42c8-af19-4992f1ee60c6" 

response = requests.get(data_url)
if response.status_code == 200:
    data_json = response.json()
    print(data_json)
    if data_json["success"]:
        registros = data_json["result"]["records"]

        # Extraer la info relevante
        parkings = [{"id": r["id"], "plazas_libres": r["libres"]} for r in registros]
        print(parkings)
        df = pd.DataFrame(parkings)
    else:
        raise Exception("Error en la respuesta de la API")
else:
    raise Exception("Error al descargar el dataset")
print(df.head())
```

Al comprobar que está todo en orden procedenos a hacer merge con el dataset que contiene los nombres de los parkings.


```python
catalogo_url = "https://datosabiertos.malaga.eu/recursos/aparcamientos/ocupappublicosmun/catalogo.csv"

# Cargar el catálogo de aparcamientos
catalogo_df = pd.read_csv(catalogo_url, sep=',', encoding='utf-8')

# Asegurar que las columnas 'id' en ambos DataFrames estén limpias y en el mismo formato
catalogo_df
```
