# Crear entorno virtual con `uv` para `mdb_ai_agents`

Pasos mínimos para instalar `uv`, crear `.venv` e instalar `requirements.txt`.

## 1. Instalar `uv`

Mac/Linux:
```bash
curl -LsSf https://astral.sh/uv/install.sh | sh
```

Windows PowerShell:
```powershell
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

````python
!uv --version
````

## 2. Entrar en el proyecto

````python
%cd <donde lo hayas clonado>/mdb_ai_agents
````

## 3. Crear el entorno virtual

````python
!uv venv .venv
````

## 4. Activar el entorno

Mac/Linux:
```bash
source .venv/bin/activate
```

Windows PowerShell:
```powershell
.venv\Scripts\Activate.ps1
```

## 5. Instalar dependencias

````python
!uv pip install -r requirements.txt
````

## 6. Verificar

Si quieres comprobar que el entorno quedó bien creado:

````python
!.venv/bin/python --version
````
